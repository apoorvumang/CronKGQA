```python
import pickle
```

```python
dataset = 'wikidata_big'
split = 'test'
filename = 'data/{dataset_name}/questions/{split}.pickle'.format(
    dataset_name=dataset,
    split=split)
questions = pickle.load(open(filename, 'rb'))
```

```python
ba_questions = []
question_type_needed = 'first_last'
for i, q in enumerate(questions):
    if q['type'] == question_type_needed:
        qc = q.copy()
        qc['id'] = i
        ba_questions.append(qc)
```

```python
import random
ids = []
num = 10
for i in range(num):
    ids.append(random.choice(ba_questions)['id'])

print('ids =', ids, '# {}'.format(question_type_needed))
```
